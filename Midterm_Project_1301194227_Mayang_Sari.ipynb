### Case 2
Buatlah plot 3 dimensi dengan menggunakan surface plot dan contour plot untuk fungsi

````python
from mpl_toolkits import mplot3d
import numpy as np
````

````python
def f(x, y):
   return (np.sin(x **2) + np.cos(y **2))
	
x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)

X, Y = np.meshgrid(x, y)
Z = f(X, Y) 
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
 
# Creating plot
surf = ax.plot_surface(X, Y, Z, cmap='Blues')

fig.colorbar(surf, ax = ax,
             shrink = 0.5, aspect = 5) 
# # show plot
ax.set_title('Surface plot')
plt.show()
````

````python
def f(x, y):
   return (np.sin(x **2) + np.cos(y **2))
	
x = np.linspace(-5, 5, 25)
y = np.linspace(-5, 5, 25)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='Purples')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
````

````python
data = pd.read_csv("AirPassengers.csv", header=0, index_col=0, parse_dates=True)
data
````

#### Problem A
Buatlah plot data time series jumlah penumpang dari tahun 1950 sampai tahun 1959.

````python
ax = data.plot(color='blue', figsize=(10, 7))
ax.set_xlim(pd.Timestamp('1950-01-01'), pd.Timestamp('1959-12-01'))

ax.set_xlabel('Years')
ax.set_ylabel('Number of passengers')
ax.set_title('Penumpang dari tahun 1950 sampai tahun 1959')
````

#### Problem B
Hitunglah kenaikan atau penurunan jumlah penumpang tiap bulannya, lalu buat plotnya.

````python
# menggunakan built-in function dari python untuk menghitung kenaikan/penurunan jumlah #Passengers
diff = data['#Passengers'].diff()
````

````python
plt.clf()

xs = data.index 
ys = diff

plt.figure(figsize=(30,10))
plt.plot(xs,ys,'bo-')

for a,b in zip(xs,ys):

    label = "{:.2f}".format(b)

    plt.annotate(label, 
                 (a,b), 
                 textcoords="offset points",
                 xytext=(0,7),
                 ha='center')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Kenaikan atau penurunan jumlah penumpang tiap bulannya')
plt.show()
````

#### Problem C
Buatlah histogram untuk menunjukkan distribusi angka kenaikan atau penurunan jumlah
penumpang.

````python
diff.plot.hist(rwidth=0.9, color='#607c8e', figsize=(10,5))
plt.title('Distribusi angka kenaikan atau penurunan jumlah penumpang')
plt.show()
````
